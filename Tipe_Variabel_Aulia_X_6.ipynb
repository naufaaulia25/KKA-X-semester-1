# Jenis Operasi Aritmatika di Pyton

```python
#penjumlahan
apel_alecia = 15
apel_silla = 20

jumlah_apel = apel_alecia + apel_silla

print(jumlah_apel)
```

```python
#pengurangan
apel_alecia = 15
apel_silla = 20

jumlah_apel = apel_silla - apel_alecia

print(jumlah_apel)
```

```python
#perkalian
jumlah_botol_air = 24
jumlah_dus_air = 3

total_botol_air = jumlah_botol_air * jumlah_dus_air

print(total_botol_air)
```

```python
#pembagian
jumlah_anak_yatim = 7

botol_tiap_anak = total_botol_air / jumlah_anak_yatim

print(botol_tiap_anak)
```

```python
#pembagian bulat
jumlah_anak_yatim =7

botol_tiap_anak = total_botol_air // jumlah_anak_yatim

print(botol_tiap_anak)
```

```python
#modulo
sisa_botol_air = total_botol_air % jumlah_anak_yatim

print(sisa_botol_air)
```

```python
sisi = 5
luas_persegi = sisi ** 2

print(luas_persegi)
```

```python
luas_persegi_1 = 36

sisi = luas_persegi_1 ** 0.5

print(sisi)
```
